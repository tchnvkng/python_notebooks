```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
plt.rcParams['figure.figsize']=[16,9]
```

```python
t=np.array([70,120,180,360])
f=np.array([10,20,35,55])

tt=np.array([7,90,180,360])
ff=np.array([50,20,15,15])

loc=(-t).argsort()
t=t[loc]
f=f[loc]
tt=tt[loc]
ff=ff[loc]


b=(f/t).cumsum()
a=b*0
a[1:]=b[0:-1]
plt.plot([0,0],[0,b.max()],color='black',linewidth=4)
plt.plot([0,t.max()],[0,0],color='black',linewidth=4)
for i in range(t.shape[0]):
    plt.plot([0,t[i]],[b[i],b[i]],color='black',linewidth=4)
    plt.plot([t[i],t[i]],[a[i],b[i]],color='black',linewidth=4)
    plt.plot([t[i],t[i]+7],[b[i],b[i]],color='black',linewidth=1)
    plt.arrow(t[i]+5,a[i],0,b[i]-a[i])
    plt.text(t[i]+7,0.5*(a[i]+b[i]),'{:d}/{:d} invested daily'.format(ff[i],tt[i]))
    plt.text(t[i]+7,(0.65*a[i]+0.35*b[i]),'with a maturity of {:d}d'.format(tt[i]))
    plt.text(t[i]/3-15,0.5*(a[i]+b[i]),'({:d}/{:d})*{:d}={:d}'.format(ff[i],tt[i],tt[i],ff[i]))
    plt.text(t[i]/3-15,(0.65*a[i]+0.35*b[i]),'outstanding after {:d}d'.format(tt[i]))
plt.xlim([0,500])
plt.ylim([0,1.1*b.max()])
plt.xticks(t,['360d','180d','90d','7d'])
plt.yticks([0,b.max()],['',''])
plt.grid('True')
#plt.axis('off')
```
